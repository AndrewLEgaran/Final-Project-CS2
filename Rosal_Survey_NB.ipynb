```python
survey_ref = db.reference("survey")
survey_data = survey_ref.get()
print(survey_data)
```

```python
respondent_avg_scores = []
question_keys = ['q1_satisfaction', 'q2_usability', 'q3_design', 'q4_clarity', 'q5_recommend', 'q6_overall']

for respondent in survey_data:
    if respondent is not None:
        scores_for_respondent = []
        for q_key in question_keys:
            if q_key in respondent and isinstance(respondent[q_key], (int, float)):
                scores_for_respondent.append(respondent[q_key])

        if scores_for_respondent:
            average_for_respondent = sum(scores_for_respondent) / len(scores_for_respondent)
            respondent_avg_scores.append({
                'respondent_id': respondent.get('respondent_id', 'N/A'),
                'average_score_q1_q6': f"{average_for_respondent:.2f}"
            })

print("Average Scores for each Respondent (Q1-Q6):")
for avg_score_data in respondent_avg_scores:
    print(f"  Respondent ID: {avg_score_data['respondent_id']}, Average Score: {avg_score_data['average_score_q1_q6']}")
```
