```python
ref = db.reference("survey")

while True:
  print("\n===== Showtime Survey Menu! =====")
  print("1. Showtime!: ") #Display
  print("2. Contestant Applications!: ") #Add new survey data
  print("3. Scene Change!: ") #Update data
  print("4. Prop Removal!: ") #Delete data
  print("5. House's Tricks! ") #Displays The Planned Features
  print("6. Grand Finale!: ") #Exit
  print("=================================")


  Decision = input("Please Enter Your Decision: ")

  # Showtime | Display
  if Decision == "1":
    survey_data_from_db = ref.get()
    print("\nStudent List: ")
    # Iterate through the list, skipping None entries (Firebase returns list with None at index 0 if keys are sequential integers)
    for survey_item in survey_data_from_db:
      if survey_item is not None:
        print(f"ID: {survey_item['respondent_id']}, Age: {survey_item['age']}, Gender: {survey_item['gender']}, Grade_Level: {survey_item['grade_level']}, Satisfaction: {survey_item['q1_satisfaction']}, Usability: {survey_item['q2_usability']}, Design: {survey_item['q3_design']}, Clarity: {survey_item['q4_clarity']}, Suggest: {survey_item['q5_recommend']}, Overall: {survey_item['q6_overall']}, Feedback: {survey_item['q7_comments']}, Faction: {survey_item['section']} ")

  #Contestant Applications! | Add new survey data
  elif Decision == "2":
    NewID = input("Enter Applicant ID: ")
    AgeVerification = input("Enter Applicant Age: ")
    GenderIdentity = input("Enter Applicant Gender: ")
    GradeIQ = input("Enter Applicant Grade Level: ")
    Satisfaction = input("Enter Satisfaction Rating (1-5): ")
    Usability = input("Enter Usability Rating (1-5): ")
    Design = input("Enter Design Rating (1-5): ")
    Clarity = input("Enter Clarity Rating (1-5): ")
    Suggest = input("Enter Recommendation Rating (1-5): ")
    Overall = input("Enter Overall Rating (1-5): ")
    Feedback = input("Enter Applicant's Commentations: ")
    Faction = input("Enter Applicant's Section: ")

    survey = {
        "respondent_id": int(NewID),
        "age": int(AgeVerification),
        "gender": GenderIdentity,
        "grade_level": GradeIQ,
        "q1_satisfaction": int(Satisfaction),
        "q2_usability": int(Usability),
        "q3_design": int(Design),
        "q4_clarity": int(Clarity),
        "q5_recommend": int(Suggest),
        "q6_overall": int(Overall),
        "q7_comments": Feedback,
        "section": Faction

    }
    ref.child(str(NewID)).set(survey)
    print("Data uploaded successfully!")

  #Scene Change | Update
  elif Decision == "3":
    UpdateID = input("Enter ID to update: ")
    # Corrected: Use UpdateID to target a specific child | <-- AI problem explanations
    target_survey_ref = ref.child(UpdateID)

    # Corrected: Get the data for the specific ID
    current_survey_data = target_survey_ref.get()

    if current_survey_data: # Check if data exists for the given ID
      NewID = (input("Enter New Applicant ID: "))
      AgeVerification = (input("Enter New Applicant Age: "))
      GenderIdentity = input("Enter New Applicant Gender: ")
      GradeIQ = (input("Enter New Applicant Grade Level: "))
      Satisfaction = (input("Enter New Satisfaction Rating (1-5): "))
      Usability = (input("Enter New Usability Rating (1-5): "))
      Design = (input("Enter New Design Rating (1-5): "))
      Clarity = (input("Enter New Clarity Rating (1-5): "))
      Suggest = (input("Enter New Recommendation Rating (1-5): "))
      Overall = (input("Enter New Overall Rating (1-5): "))
      Feedback = input("Enter New Applicant's Commentations: ")
      Faction = input("Enter New Applicant's Section: ")

      target_survey_ref.update({
        "respondent_id": int(NewID),
        "age": int(AgeVerification),
        "gender": GenderIdentity,
        "grade_level": GradeIQ,
        "q1_satisfaction": int(Satisfaction),
        "q2_usability": int(Usability),
        "q3_design": int(Design),
        "q4_clarity": int(Clarity),
        "q5_recommend": int(Suggest),
        "q6_overall": int(Overall),
        "q7_comments": Feedback,
        "section": Faction
      })

      print("Student Info Has been succcessfully updated.")

    else:
      print(f"Student with ID {UpdateID} cannot be found within the list.")

  #Prop Removal | Delete
  elif Decision == "4":
      survey_db = input("Please enter ID of the Student You want to Erase from existence within the list. ")

      if ref.child(survey_db).get():
          ref.child(survey_db).delete()
          print("Student Info Has been succcessfully deleted.")
      else:
        print("Student could not be found, Requiring Information of the Student.")

  #House's Tricks! | Features
  elif Decision == "5":
      print("House's Tricks!")
      print("1. Average of each question from q1 to q6:")
      print("2. Question with the highest average rating:")
      print("3. Question with the lowest average rating:")
      print("4. Displays Comments with suggestions of improvement:")
      print("5. Displays Comments with satisfied results:")

      feature_db = input("Please Enter The Number of The Feature you want to be implemented: 1-5 ")

      if feature_db == "1":
        keys_to_average = ['q1_satisfaction', 'q2_usability', 'q3_design', 'q4_clarity', 'q5_recommend', 'q6_overall']

        print("Averages for selected keys:")
        for key in keys_to_average:
          total_sum = 0
          count = 0
          # Re-fetch or ensure survey_data is accessible here if needed for this loop
          survey_data_from_db = ref.get()
          for record in survey_data_from_db:
            if record is not None and key in record and isinstance(record[key], (int, float)):
              total_sum += record[key]
              count += 1

          if count > 0:
            average = total_sum / count
            print(f"  The average for '{key}' is: {average:.2f}")
          else:
            print(f"  No valid data found for '{key}' to calculate the average.")

      elif feature_db == "2":
        keys_to_compare = ['q1_satisfaction', 'q2_usability', 'q3_design', 'q4_clarity', 'q5_recommend', 'q6_overall']
        question_averages = {}
        survey_data_from_db = ref.get()

        for key in keys_to_compare:
            total_sum = 0
            count = 0
            for record in survey_data_from_db:
                if record is not None and key in record and isinstance(record[key], (int, float)):
                    total_sum += record[key]
                    count += 1
            if count > 0:
                question_averages[key] = total_sum / count
            else:
                question_averages[key] = 0 # Assign 0 if Data is invalid

        if question_averages:
            highest_avg_question = max(question_averages, key=question_averages.get)
            highest_avg_value = question_averages[highest_avg_question]
            print(f"\nQuestion with the highest average score (Q1-Q6): '{highest_avg_question}' with an average of {highest_avg_value:.2f}")
        else:
            print("Glitch, No Question is compared.")

      elif feature_db == "3":
        keys_to_compare = ['q1_satisfaction', 'q2_usability', 'q3_design', 'q4_clarity', 'q5_recommend', 'q6_overall']
        question_averages = {}
        survey_data_from_db = ref.get()

        for key in keys_to_compare:
            total_sum = 0
            count = 0
            for record in survey_data_from_db:
                if record is not None and key in record and isinstance(record[key], (int, float)):
                    total_sum += record[key]
                    count += 1
            if count > 0:
                question_averages[key] = total_sum / count
            else:
                question_averages[key] = float('inf') # Assign infinity if no valid data, to ensure it's not picked as lowest

        if question_averages and any(value != float('inf') for value in question_averages.values()):
            lowest_avg_question = min(question_averages, key=question_averages.get)
            lowest_avg_value = question_averages[lowest_avg_question]
            print(f"\nQuestion with the lowest average score (Q1-Q6): '{lowest_avg_question}' with an average of {lowest_avg_value:.2f}")
        else:
            print("Glitch, No valid questions to compare averages.")

      elif feature_db == "4":
        print("Comments with suggestions of improvement:")
        survey_data_from_db = ref.get()
        improvement_keywords = ['improve', 'needs', 'should be', 'could be', 'suggestion', 'suggestions', 'better', 'faster', 'clearer', 'confusing']
        found_suggestions = False
        for survey_item in survey_data_from_db:
          if survey_item is not None and 'q7_comments' in survey_item and survey_item['q7_comments']:
            comment = survey_item['q7_comments'].lower()
            if any(keyword in comment for keyword in improvement_keywords):
              print(f"  - ID {survey_item['respondent_id']}: {survey_item['q7_comments']}")
              found_suggestions = True
        if not found_suggestions:
            print("  No comments with explicit suggestions for improvement found.")

      elif feature_db == "5":
        print("Comments with satisfied results:")
        survey_data_from_db = ref.get()
        satisfied_keywords = ['satisfied', 'happy', 'great', 'excellent', 'loved', 'perfect', 'enjoyed', 'good', 'solid', 'smooth', 'engaging', 'helpful', 'ease of use', 'recommend']
        found_satisfied = False
        for survey_item in survey_data_from_db:
          if survey_item is not None and 'q7_comments' in survey_item and survey_item['q7_comments']:
            comment = survey_item['q7_comments'].lower()
            if any(keyword in comment for keyword in satisfied_keywords):
              print(f"  - ID {survey_item['respondent_id']}: {survey_item['q7_comments']}")
              found_satisfied = True
        if not found_satisfied:
            print("  No comments indicating satisfaction found.")

  #Grand Finale | Exit
  elif Decision == "6":
    print("Thank you for using the Showtime Survey Menu!")
    break

  #Error 404
  else:
    print("Error 400, Please try again...")
```

```python
survey_ref = db.reference("survey")
survey_data = survey_ref.get()

respondent_avg_scores = []
question_keys = ['q1_satisfaction', 'q2_usability', 'q3_design', 'q4_clarity', 'q5_recommend', 'q6_overall']

for respondent in survey_data:
    if respondent is not None:
        scores_for_respondent = []
        for q_key in question_keys:
            if q_key in respondent and isinstance(respondent[q_key], (int, float)):
                scores_for_respondent.append(respondent[q_key])

        if scores_for_respondent:
            average_for_respondent = sum(scores_for_respondent) / len(scores_for_respondent)
            respondent_avg_scores.append({
                'respondent_id': respondent.get('respondent_id', 'N/A'),
                'average_score_q1_q6': f"{average_for_respondent:.2f}"
            })

print("Average Scores for each Respondent (Q1-Q6):")
for avg_score_data in respondent_avg_scores:
    print(f"  Respondent ID: {avg_score_data['respondent_id']}, Average Score: {avg_score_data['average_score_q1_q6']}")
```
